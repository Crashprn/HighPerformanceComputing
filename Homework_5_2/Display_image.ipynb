```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
cpu_file_name = 'transposed_image_cpu.raw'
gpu_file_name = 'transposed_image_gpu.raw'

cpu = np.fromfile(cpu_file_name, dtype=np.uint8, sep="")
cpu = cpu.reshape((1024, 1024, 3))

gpu = np.fromfile(gpu_file_name, dtype=np.uint8, sep="")
gpu = gpu.reshape((1024, 1024, 3))

fig, axs = plt.subplots(1, 2)
axs[0].imshow(cpu)
axs[0].set_title('CPU')
axs[1].imshow(gpu)
axs[1].set_title('GPU')
plt.show()
```
