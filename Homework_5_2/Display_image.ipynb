```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
untransposed_file_name = 'gc_conv_1024x1024.raw'
cpu_file_name = 'transposed_image_cpu.raw'
gpu_reg_file_name = 'transposed_image_gpu_reg.raw'
gpu_tiled_file_name = 'transposed_image_gpu_tile.raw'

untransposed = np.fromfile(untransposed_file_name, dtype=np.uint8, sep="")
untransposed = untransposed.reshape((1024, 1024, 3))

cpu = np.fromfile(cpu_file_name, dtype=np.uint8, sep="")
cpu = cpu.reshape((1024, 1024, 3))

gpu_reg = np.fromfile(gpu_reg_file_name, dtype=np.uint8, sep="")
gpu_reg = gpu_reg.reshape((1024, 1024, 3))

gpu_tiled = np.fromfile(gpu_tiled_file_name, dtype=np.uint8, sep="")
gpu_tiled = gpu_tiled.reshape((1024, 1024, 3))

fig, axs = plt.subplots(1, 4)
axs[0].imshow(untransposed)
axs[0].set_title('Untransposed')
axs[1].imshow(cpu)
axs[1].set_title('CPU')
axs[2].imshow(gpu_reg)
axs[2].set_title('GPU Reg')
axs[3].imshow(gpu_tiled)
axs[3].set_title('GPU Tiled')

plt.show()
```
